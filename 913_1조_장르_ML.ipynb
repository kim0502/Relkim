```r
# sampling

sampling <- function(){
samp <<- createDataPartition(genre3$흥행여부, p = 0.7,list = F) 
data.tr <<- genre3[samp,]                                     
data.test <<- genre3[-samp,]

x <<- subset(data.test, select = - 흥행여부)
y <<- data.test$흥행여부
    
}  
```

```r
sampling()
head(samp)
```

```r
#인공신경망 테스트
ANN <- function(data.tr , data.test , x , y ){
    model.nnet <- nnet(흥행여부 ~. , data = data.tr , size = 100)
    pred <- predict(model.nnet , x , type = "class")
    cat("예측력" ,mean(pred == y), "입니다.")
    table(pred,y)
    return(mean(pred == y))
}

#의사결정 나무 
CTREE <-function(data.tr , data.test , x , y ){
    model.ctree <- ctree(흥행여부 ~ . , data=data.tr)
    pred <- predict(model.ctree , x )
    cat("예측력" ,mean(pred == y), "입니다.")
    table(pred,y)
    return(mean(pred == y))
}

#RandomForesst
RandomForest<-function(data.tr , data.test , x , y ){
    model.rf <- randomForest(흥행여부 ~ . , data=data.tr, ntree = 100)
    pred <- predict(model.rf , x )
    cat("예측력" ,mean(pred == y), "입니다.")
    table(pred,y)
    return(mean(pred == y))
} 

#SVM
SVM <-function(data.tr , data.test , x , y ){
    model.svm <- svm(흥행여부 ~ . , data=data.tr)
    pred <- predict(model.svm , x )
    cat("예측력" ,mean(pred == y), "입니다.")
    table(pred,y)
    return(mean(pred == y))
} 

#Logistic regression
Logisticreg <-function(data.tr , data.test , x , y ){
    model.log <- multinom(흥행여부 ~ . , data=data.tr)
    pred <- predict(model.log , x )
    cat("예측력" ,mean(pred == y), "입니다.")
    table(pred,y)
    return(mean(pred == y))
} 
```

```r
 ANN(data.tr , data.test , x , y )
 CTREE(data.tr , data.test , x , y )
```

```r
set.seed(1234)
a <- rep(0,5)
iteration <- 5
for(i in 1:iteration){
    
    a = a + c(
            ANN(data.tr , data.test , x , y ),
            CTREE(data.tr , data.test , x , y ),
            RandomForest(data.tr , data.test , x , y ),
            SVM(data.tr , data.test , x , y ),
            Logisticreg(data.tr , data.test , x , y )
    )
    
}
a<- a/iteration

result<-data.frame(예측력 = a)
rownames(result) = c("인공신경망","의사결정나무","RandomForest","SVM","로지스틱회귀분석")
```
