### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
plot(price)
```

```r
# 2행2열 차트 그리기 : type 속성으로 구분해 보기
#12
#34
par(mfrow = c(2,2))
plot(price, type='l')
plot(price, type='o')
plot(price, type='h')
plot(price, type='s')
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기

par(mfrow = c(2,2))
plot(price, type="o",pch=5)
plot(price, type="o",pch=15)
plot(price, type="o",pch=20, col='blue')
plot(price, type="o",pch=25, col='orange')
plot(price, type="o",pch=c(rep(15,8),20), col='orange')
plot(price, type="o",pch=c(rep(15,8),20), col='orange',cex=c(rep(1,3),3,1))
```
