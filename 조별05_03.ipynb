```r
이 <- (election$이재명/election$투표수)*100
윤 <- (election$윤석열/election$투표수)*100
심 <- (election$심상정/election$투표수)*100


election$기타 <-100-(이+윤+심)
election
```

```r
bus <- aggregate(cbind(투표수,선거인수,이재명,윤석열,심상정,기타)~광역시도,election,sum)
bus
```

```r
bus$투표율 <- bus$투표수/bus$선거인수*100
bus$이재명 <- bus$이재명/bus$투표수*100
bus$윤석열 <- bus$윤석열/bus$투표수*100
bus$심상정 <- bus$심상정/bus$투표수*100
bus1<-arrange(bus,desc(투표율))
bus1
```

```r
options(scipen = 2)
options(warn = -1)
par(family = "AppleGothic")

barplot(bus3,
      main ="투표율 최고,최저 지역의 득표율",
      ylab = "투표율 = 지역별 투표수/선거인수",
      xlab = "후보별 득표율(%)",
      xlim = c(0,100),  
      col = c("skyblue","red","yellow","grey"),
      cex.names = 1,  
      horiz = T,
      legend = T,
      beside = T,
)
```
