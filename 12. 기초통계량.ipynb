### 기본 통계량으로 분석 해보기 (기본통계량=평균)

```r
# 밀가루 사용

flour <-c(3, -2, -1, 0, 1, -2)

#다이어트 약
dite <- c(-4, 1, -3, -5, -2, -8)
```

```r
total <- c(flour,dite)
total
```

```r
# 기본 통계량

summary(total)
```

```r
# 히스토그램
hist(total)
```

```r
boxplot(total)
```

```r
boxplot(flour,dite)
```

```r
#상관계수 : 두 데이터사이에 관계를 알아볼때 ,어떤기준에서 다른것과 관계가있는지 알아볼때
cor(flour,dite) #결과 값을 보면 두 사이에는 상관없다.
```

```r
x<-c(1,3,9)   # 두값의 위치가 결과값을 내는데 있어 중요하다.
y<-c(10,20,30)
```

```r
par(mfrow = c(1,2))
plot(x, type="l")
plot(y, type="l")
```

```r
cor(x,y)
```

```r
cafe <-read.csv("../../Data/coffee.csv")
head(cafe)
```

```r
# 커피의 빈도수  
#빈도수는 table로 한다.
table(cafe$Coffee)
```
