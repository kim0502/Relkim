# 선그래프

```r
# 분석할때,해석이 중요하다(command가 아닌 데이터가 중요하다.)
# 데이터를 한 그래프에서 비교할 수 있는 그래프
```

```r
# 3 종류 dataset의 선 그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)

#같은 cell에서 한번에 해야한다.
```

```r
#par(family = "AppleGothic")
plot(abc,
    type = "o",
    col = "red", #y축에 범위조정이 되어있지 않아 ylim을 추가해줘야한다. 비교하려면 ylim을 지워보면 안다.
     ylim = c(0, 400),
     axes = F, #x,y 좌표 Frame 제거
     ann = F # x,y의 label 제거
    )

# x축과 x축의 라벨을 설정
axis(1, at=1:5, lab = c('서울','대전','대구','광주','원주')) # 축 만드는 함수이다. 영어로 단수 이다.

# y축과 y축의 라벨을 설정
axis(2, ylim=c(0,400))

# main title 지정
title(main='과일 판매량', col.main = "red")

# x축 title(label) 지정
title(xlab = "지역", col.lab ="blue") # x로만 적어도 된다.

# y축 title(label) 지정
title(ylab = "판매량", col.lab ="green") # y로만 적어도 된다.

# 선그리기 함수로 def 그리기
lines(def,
     type ="o",
     pch = 22,
     col = "green",
     lty = 2)

# 선그리기 함수로 ghi 그리기
lines(ghi,
      type = 'o',
      pch = 22,
      col = "blue",
      lty = 2
      )

# 범례 아무곳이나 쓸수 있는함수
legend(4,400, # x,y좌표 값
        c("야구장","축구장","해변가"),
        col = c("red","green","blue"),
        pch = c(21,22,22),
       lty =  c(1, 2, 2),
       cex = 0.8 # 기본값이 무조건 1이다.
      )
```

```r
# par(family = 'AppleGothic')  

sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
retio <- sales /sum(sales)*100
label <- paste(week,"\n",retio,"%")
```

```r
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
retio <- sales /sum(sales)*100
label <- paste(week,"\n",retio,"%")

# par(family = 'AppleGothic') 
pie3D(sales,
   main = "영업실적 파이차트",
   #labels = week
   labels = label,
    explode = 0.05,
    labelcex = 0.8
)
```
