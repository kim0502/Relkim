```r
x <- 10
y <- runif(1) * 100

if(x > y){
    print("x > runif()")
}
print("---End---")
```

```r
x <- 10
y <- runif(1) * 100

if(x > y){
    print("x > runif()")
}else{
    print("x > runif()")
}
print("---End---")
```

```r
x <- 10
y <- runif(1) * 100

if(x > y){
    print("x > runif()")
}else if(x < y){
    print("x < runif()")
}else{
    print("x = runif()")
}
```

```r
# 홀수 짝수 구분

a <- c(8,7,6)

if(a%%2==0){
    sprintf("%i는 짝수 입니다.",a)
}else{
    cat(a,"는 홀수 입니다.")
}
```

```r
a <- c(8,7,6)

ifelse(a %% 2 == 0, "짝수", "홀수")
```

```r
ifelse(a %% 2 == 0, paste(a,": 짝수"), paste(a,": 홀수")) #벡터로 만들기위해 paste로 해준다.
```

```r
# 사용자로부터 data를 입력 받아 입력한 숫자가 홀수인지 짝수 인지 판단하여 출력하기.

num1 <- as.integer(readline("숫자를 입력하세요 :"))

#입력하신 숫자는 12이고 짝수 입니다.

if(num1 %% 2 ==0){
  cat("입력하신 숫자는",num1,"이고 짝수 입니다.") 
}else{ 
    cat("입력하신 숫자는",num1,"이고 홀수 입니다.") 
}
```

```r
a <- as.integer(readline("점수를 입력하세요 :"))

if(a >= 90){
    print("A학점 입니다.")
}else if( a >= 80){
    print("B학점 입니다.")
}else{
    print("C학점 입니다.")
}
    
  
```

### csv file을 불러서 사용하기

```r
excel <- read.csv("../../Data/excel.csv")
head(excel)
```

```r
# q1컬럼에서 Data가 3보다 크거나 같으면 data를 제곱근을 구하고 아니면 원래 Data를 출력하기

q1 <- excel$q1
length(q1)
print(ifelse(q1>=3, sqrt(q1), q1)) #(조건, 맞으면, 틀리면)
```

```r
# q1컬럼에서 Data가 3보다 크거나 같으면 data를 제곱근을 구하고 아니면 원래 Data를 q6라는 컬럼을 만들어 추가하기-----------------------------

excel$q6 <- ifelse(q1>=3, sqrt(q1), q1) 
head(excel)
```
