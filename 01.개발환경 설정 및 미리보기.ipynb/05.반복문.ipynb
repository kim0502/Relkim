```r
# excel에서 q3 컬럼 데이터의 홀수의 합과 짝수의 합 구하기




#'홀수의 합은 604 이고 짝수의 합은 852 입니다.'
sumEven <- 0; sumOdd <- 0

for(i in excel$q3){
    if(i %% 2 ==0){
        sumEven = sumEven + i
    }else{
        sumOdd = sumOdd + i
    }
}
cat("홀수의 합은", sumOdd, "이고 짝수의 합은", sumEven, "입니다.")
```

```r
# excel에서 q3 컬럼 데이터중 3의 배수의 합계를 구하기

sum <- 0;

for(i in excel$q3){
    if(i %% 3 ==0){
        sum = sum + i
    }
} 
sum
```

```r
# excel에서 q3컬럼 데이터중 3의 배수도 되고 4의 배수도 되는 수의 합계를 구하시오.

sum <- 0;

for(i in excel$q3){
    if(i %% 3 ==0 && i %% 4 ==0){
        sum = sum + i
    }
} 
sum
```

```r
# 사용자로 부터 몇단을 출력할지 단수를 입력받아 구구단 출력하기

dan <- as.integer(readline("단을 입력하세요!"))
for(i in 1:9){
    cat(dan, "X", i, "=", dan*i, "\n")
}
```

```r
# 구구단 함수 만들기 -2
gugu <- function(x){
    for(i in 1:9){
        cat(x,"X", i, "=", x*i, "\n")
    }
}
```

```r
# 구구단 함수 만들기 -3

gugu <- function(x,y){
    for(i in 1:y){
        cat(x,"X", i, "=", x*i, "\n")
    }
}
```

### while

```r
i <- 0
while(i < 10){
    i = i + 1
    print(i)
}
```

### repeat

```r
i <- 0
repeat{
    i = i + 1
    if(i > 10){
        break
    }
    print(i)
}
```
