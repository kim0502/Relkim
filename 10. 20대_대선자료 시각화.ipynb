```r
# 데이터 불러오기
election <- read.csv("../Data/election_2022.csv", fileEncoding="utf-8" )
```

```r
str(election)
```

```r
summary(election)
```

```r
# electionSum = 이재명 + 윤석열 + 심상정
election$electionSum <- apply(election[,5:7],1,sum)
head(election)
```

```r
# electionRatio <- 특표수 대비 electionSum의 퍼센트 구해서 넣기
election$electionRatio <- election$electionSum/election$투표수 *100
head(election)
```

```r
# 3명의 후보를 가장 많이 선택한 지역과 가장 적게 선택한 지역 구하기
# 1)정렬
library(dplyr)
```

```r
arrange(election, desc(electionRatio))[1,]
arrange(election, desc(electionRatio))[nrow(election),]
```

```r
# 2) subset
subset (election, electionRatio == max(election$electionRatio))
subset (election, electionRatio == min(election$electionRatio))
```

```r
# election_2022_job1.csv로 저장
write.csv(election, "../Data/election_2022_job1.csv", row.names =F, fileEncoding="utf-8")
```

```r
# 각 후보(윤석열, 이재명, 심상정)의 광역시도 별 득표율을 3D Pie Chart로 표현하기
# (단, 득표율이 3% 미만인 광역시도는 광역시도 명만 출력하고 퍼센트는 출력하지 않는다. / 페센트 단위는 소수 한자리)
```

```r
vote3<- aggregate(심상정~광역시도,election,sum)
vote3
```

```r
vote4 <- arrange(vote3,desc(심상정)) 
vote4
```

```r
options(warn = -1)
심상정 <- vote4$심상정
광역시도 <- vote4$광역시도
득표 <- round((심상정 /sum(심상정)*100),1)
label <- ifelse(득표>3,paste(광역시도,"\n",득표,"%"),as.character(광역시도))

par(family = 'AppleGothic') 

pie3D(심상정,
   labels = label,
   main = "심상정 후보의 광역시도별 득표율",
   col = rainbow(length(득표)),
  labelcex = 0.7
  
)
```

```r
# 수도권(경기도, 서울, 인천)의 투표인구 수의 퍼센트는?
vote0<- aggregate(투표수~광역시도,election,sum)
vote0
vote01<-round(sum(c(vote0[2,2],vote0[9,2],vote0[12,2]))/sum(투표수)*100,2)
vote01
```

# 투표수가 가장 많은 4지역(광역시도)의 후보별 차이를 막대그래프로 시각화

```r
str(election)
```

```r
subset(election,광역시도 ==c('경기도','서울특별시', '부산광역시','경상남도'))
most <- aggregate(cbind(투표수,이재명,윤석열,심상정)~광역시도,election,sum)
most
most1 <- arrange(most,desc(투표수))
most1
```

```r
most2<-most1[c(1:4),c(3:5)]
most2
most3<-t(most2)
most3
```

```r
colnames(most3) <- c("경기도","서울특별시","부산광역시","경상남도")
most3
```

```r
options(scipen = 2)
options(warn = -1)
par(family = "AppleGothic")

barplot(most3,
     beside = T,
     main = "투표수가 가장 많은 4지역의 후보별 득표수 차이" ,
     ylim = c(0,5000000),
     col = c("skyblue","red","yellow"), 
     legend.text = T,    
        
)
```

# 투표수가 높은 시군구의 후보별 득표수를 막대그래프로 시각화

```r
head(election)
```

```r
home<-arrange(election,desc(투표수))
head(home)
```

```r
home1<-home[1:5,5:7] #matrix 해주기
home1
```

```r
home2 <- t(home1) #matrix 해주기
home2
```

```r
colnames(home2) <- c ("경기도 화성시","경기도 부천시","경기도 남양주시","서울특별시 송파구","서울특별시 강서구")
home2
```

```r
options(scipen = 100)
options(warn = -1)
par(family = "AppleGothic")

barplot(home2,
        main = "투표수 높은 지역의 후보별 투표수",
        names = c("경기도 화성시","경기도 부천시","경기도 남양주시","서울특별시 송파구","서울특별시 강서구"),
        ylim = c(0,600000),
        legend.text = T, 
        col = rep(c("skyblue","red","yellow")),
        cex.names = 0.6
        
)
```
