```r
seoul <- read.table("../Data/trafficAccident_2011_2019.txt", header = T, stringsAsFactors = F)
head(seoul)
```

```r
tail(seoul)
```

```r
# 합계 column 제외하기
seoul <- subset(seoul, 자치구별 != "합계")
seoul <- subset(seoul, select = -합계)
```

```r
# 숫자들 중에서 -인 것들 0으로 바꾸기
for (i in 4:15) {
    seoul[seoul[,i] == "-",i] <- 0
    seoul[,i] <- as.numeric(seoul[,i])
}
```

```r
seoul
```

```r
# 각 컬럼들에 해당하는 데이터 뽑아오기

# seoul의 row를 3개씩 나눠서 생각하기
num <- seq(1,nrow(seoul), 3)

# 발생건수, 사망자수, 부상자수 
occur <- c()
death <- c()
injure <- c()
for (i in num) {
    occur <- append(occur, as.numeric(seoul[i,4:15]))
    death <- append(death, as.numeric(seoul[i + 1,4:15]))
    injure <- append(injure, as.numeric(seoul[i + 2,4:15]))
}

# 자치구, 년도, 월 뽑기
jachigu <- rep(seoul[num,2],each=12)
year <- as.numeric(rep(seoul[num,1], each = 12))
month <- rep(1:12,length(num))

# 하나의 데이터 프레임으로 만들기
seoul <- data.frame(년도=year,월 = month,자치구별=jachigu, 발생건수 = occur, 사망자수 = death, 부상자수 = injure)
```
