```r
set.seed(1234)
samp<-c(sample(1:50,35), sample(51:100,35), sample(101:150,35))
samp
```

```r
# 학습용 데이터셋
data.tr<- iris2[samp,]
summary(data.tr)
```

```r
# 데스트용 데이터셋
data.test <- iris2[-samp,]
summary(data.test)
```

```r
# 인공신경망으로 합습시키기
library(nnet)
```

```r
# 모델만들기
model.nnet<-nnet(Species ~.,data=data.tr, size=3)
```

```r
# 문제 및 정답 만들기
x <- subset(data.test, select = -Species)
y <- data.test$Species
```

```r
# Test
pred <- predict(model.nnet,x,type="class")
head(pred)
```

```r
# 예측력 확인
mean(pred ==y)
```

```r
# 정오표 만들기
table(pred, y)
```
